# Grid 并行多图

```python
import pyecharts.options as opts
```

```python
x_data = ['Apple', 'Huawei', 'Xiaomi', 'Oppo', 'Vivo', 'Meizu']
y_data = [123, 153, 89, 107, 98, 23]
bar = (Bar()
        .add_xaxis(x_data)
        .add_yaxis('', y_data)
      )
line = (Line()
        .add_xaxis(x_data)
        .add_yaxis('', y_data)
       )
grid = (Grid()
           .add(bar, grid_opts=opts.GridOpts(pos_bottom="70%", pos_left="50%"))
           .add(line, grid_opts=opts.GridOpts(pos_left="15%"))
       )
grid.render_notebook()
```

```python
x_data = ['Apple', 'Huawei', 'Xiaomi', 'Oppo', 'Vivo', 'Meizu']
y_data = [123, 153, 89, 107, 98, 23]

bar = (Bar()
        .add_xaxis(x_data)
        .add_yaxis('', y_data)
        .set_series_opts(
          markline_opts=opts.MarkLineOpts(
              data=[
                  opts.MarkLineItem(type_="max", name="最大值"),
                  opts.MarkLineItem(type_="min", name="最小值"),
                  opts.MarkLineItem(type_="average", name="平均值")

              ]
          )
        )
      )
bar.render_notebook()
```

### 标记区域

#### 区域范围

```python
date_list = ['10月{}日'.format(i) for i in range(1, 32)]
data = [random.randint(10, 100) for _ in date_list]

line = (Line()
         .add_xaxis(date_list)
         .add_yaxis('', data)
         .set_series_opts(markarea_opts=opts.MarkAreaOpts(
             data=[
                 opts.MarkAreaItem(name="十一黄金周", x=("10月1日", "10月7日"))
             ]
         ))
       )

line.render_notebook()
```
